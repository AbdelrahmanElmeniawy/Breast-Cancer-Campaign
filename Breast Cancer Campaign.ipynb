```python
import pandas as pd
data = pd.read_csv("data.csv")
data
```

```python
data.isnull().sum()
```

```python
#Removing unimportant data and missing values 
data = data.iloc[:, 1:32]
data
```

```python
data.describe()
```

```python
#Scalling and standrization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
for i in range(1, 31):
    data.iloc[:, i:i+1] = scaler.fit_transform(data.iloc[:, i:i+1])
```

```python
data.describe()
```

```python
data
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['diagnosis']= encoder.fit_transform(data['diagnosis'])
```
